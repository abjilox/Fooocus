```python
!apt -y update -qq
!apt -y install -qq aria2

%pip install torchsde==0.2.5 einops==0.4.1 transformers==4.30.2 safetensors==0.3.1 accelerate==0.21.0
%pip install pytorch_lightning==1.9.4 omegaconf==2.2.3 gradio==3.39.0 xformers==0.0.20 triton==2.0.0 pygit2==1.12.2

import os
import re

class FooocusColab:
    class Logger:
        def __init__(self):
            pass

        def log(self, message):
            print(message)

    class Config:
        def __init__(self):
            self.use_google_drive = False #@param {type:"boolean"}
            self.share_gradio = True #@param {type:"boolean"}
            self.commit_hash = '46a6cfe' #@param {type:"string"}
            self.username = 'a' #@param {type:"string"}
            self.password = 'a' #@param {type:"string"}
            self.ngrok_token = '' #@param {type: "string"}
            self.install_dir = 'Fooocus' #@param {type: "string"}
            #self.root_dir = '~/workspace/stablediffusion/tmp'
            self.paths = FooocusColab.Paths.paths_factory(self)
            self.git_repo = 'https://github.com/lllyasviel/Fooocus'

        @property
        def model_urls(self):
            # @markdown ---
            # @markdown ### Model URLs
            # @markdown    
            # @markdown Add URLs to download models from, separated by semicolons.   
            extra_models = '' #@param {type: "string"} 

            # @markdown 
            #
            # @markdown Download base model from huggingface   
            load_base_model = True #@param {type:"boolean"}

            # @markdown 
            #
            # @markdown Download ['Dreamshaper XL'](https://civitai.com/models/112902/dreamshaper-xl10)
            load_dreamshaper = False #@param {type:"boolean"}

            # @markdown 
            #
            # @markdown Download ['SDVN6-RealXL'](https://civitai.com/models/118114)
            load_sdvn6 = False #@param {type:"boolean"}

            # @markdown 
            #
            # @markdown Download ['Copax TimeLessXL - SDXL1.0 v5'](https://civitai.com/models/118111)
            load_copax = False #@param {type:"boolean"}

            models = re.split(r'\s*[;,:.]\s*', extra_models)
            if load_base_model:
                models.prepend('https://huggingface.co/ckpt/sd_xl_base_1.0/resolve/main/sd_xl_base_1.0_0.9vae.safetensors')
            models.prepend('https://huggingface.co/ckpt/sd_xl_base_1.0/resolve/main/sd_xl_refiner_1.0_0.9vae.safetensors')
            if load_dreamshaper:
                models.append('https://civitai.com/api/download/models/126688')
            if load_sdvn6:
                models.append('https://civitai.com/api/download/models/134461')
            if load_copax:
                models.append('https://civitai.com/api/download/models/150238')   
            return models

        @property
        def lora_urls(self):
            # @markdown ---
            # @markdown ### Lora URLs
            # @markdown 
            # @markdown Add URLs to download loras from, separated by semicolons.
            extra_loras = '' #@param {type: "string"}
            
            # @markdown Download example LoRA from huggingface
            load_base_lora = True #@param {type:"boolean"}
            # https://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/resolve/main/sd_xl_offset_example-lora_1.0.safetensors
            
            # @markdown 
            #
            # @markdown Download LoRA ['better faces sdxl' - 'face enhancer v3.0'](https://civitai.com/models/119376/better-faces-sdxl)   
            #
            # @markdown No activation needed in prompt
            load_better_faces_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/142718

            # @markdown 
            #
            # @markdown Download LoRA ['perfect eyes'](https://civitai.com/models/118427/perfect-eyes-xl)   
            #
            # @markdown Activation: `green eyes`, `blue eyes`, `brown eyes`, `close up`, `perfecteyes`
            load_perfect_eyes_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/128461     
            
            # @markdown 
            #
            # @markdown Download LoRA ['Detail Tweaker XL v1.0'](https://civitai.com/122359/detail-tweaker-xl)   
            #
            # @markdown Activation: None
            load_detail_tweaker_lora = False #@param {type:"boolean"}     
            # https://civitai.com/api/download/models/135867

            # @markdown 
            #
            # @markdown Download LoRA ['NurLens Photography master. SDXL 1.0'](https://civitai.com/126442/nurlens-photography-master-sdxl-10)   
            #
            # @markdown Activation: `NurLens`, `photography`
            load_photography_master_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/138262

            # @markdown 
            #
            # @markdown Download LoRA ['Depth of Field Slider - LoRA'](https://civitai.com/135380/depth-of-field-slider-lora)   
            #
            # @markdown Activation: None
            load_depth_of_field_slider_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/149218

            # @markdown 
            #
            # @markdown Download LoRA ['Blacklight Makeup â€” SDXL LoRA'](https://civitai.com/134643/blacklight-makeup-sdxl-lora)   
            #
            # @markdown Activation: `with blacklight makeup`
            load_blacklight_makup_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/148330

            # @markdown 
            #
            # @markdown Download LoRA ['Frank Frazetta SDXL 1.0 art style lora'](https://civitai.com/119461/frank-frazetta-sdxl-10-art-style-lora)   
            #
            # @markdown Activation: `fr4z3tt4 `
            load_frank_frazetta_lora = False #@param {type:"boolean"}
            # https://civitai.com/api/download/models/129786

            loras = re.split(r'\s*[;,:.]\s*', extra_loras)
            if load_base_lora:
                loras.prepend('ttps://huggingface.co/stabilityai/stable-diffusion-xl-base-1.0/resolve/main/sd_xl_offset_example-lora_1.0.safetensors')
            if load_better_faces_lora:
                loras.append('https://civitai.com/api/download/models/142718')
            if load_perfect_eyes_lora:
                loras.append('https://civitai.com/api/download/models/128461')
            if load_detail_tweaker_lora:
                loras.append('https://civitai.com/api/download/models/135867')
            if load_photography_master_lora:
                loras.append('https://civitai.com/api/download/models/138262')
            if load_depth_of_field_slider_lora:
                loras.append('https://civitai.com/api/download/models/149218')
            if load_blacklight_makup_lora:
                loras.append('https://civitai.com/api/download/models/148330')
            if load_frank_frazetta_lora:
                loras.append('https://civitai.com/api/download/models/129786')

            return loras                      

    class Paths:
        @classmethod
        def paths_factory(cls, config):
            if config.use_google_drive:
                return cls.GoogleDrive(config)
            else:
                return cls.Local(config)
            
        class Base:
            def __init__(self, config):
                self.config = config

            @property    
            def repo_dir(self):
                return f'{self.mount_point}/{self.config.install_dir}' 
            
            @property
            def static_dir(self):
                return self.repo_dir

            @property
            def lib_dir(self):
                return f'{self.static_dir}/lib'

            @property
            def model_dir(self):
                return f'{self.static_dir}/models'

            @property
            def checkpoint_dir(self):
                return f'{self.model_dir}/checkpoints'

            @property
            def lora_dir(self):
                return f'{self.model_dir}/loras'   
                
        class GoogleDrive(Base):
            def __init__(self, config):
                super().__init__(config)
                self.mount_dir = '/content/drive'
                self.mount_point = '/content/drive/MyDrive'   

        class Local(Base):
            def __init__(self, config):
                super().__init__(config)
                self.mount_dir = '/content'
                self.mount_point = '/content'         

    class Downloader:
        @classmethod
        def download(cls):
           downloader = cls()
           downloader.download_models()
           downloader.download_loras()

        def __init__(self):
            self.config = FooocusColab.Config()
            self.log = FooocusColab.Logger().log

        def download_models(self):
            self.log('Downloading models')
            self.load_files_from_url(url=self.config.model_urls, target_dir=self.config.paths.checkpoint_dir)
            self.log('Downloaded models') 

        def download_loras(self):
            self.log('Downloading loras')
            self.load_files_from_url(url=self.config.lora_urls, target_dir=self.config.paths.lora_dir)
            self.log('Downloaded loras')

        def load_files_from_url(self, url, target_dir):
            for url in url:
                !aria2c --console-log-level=error -c -x 16 -s 16 -k 1M {url} -d {target_dir}
                self.log(f'Downloaded {url}')

    class Setup:

        def __init__(self):
            self.config = FooocusColab.Config()
            self.log = FooocusColab.Logger().log

        def clone_repo_if_needed(self):
            if os.path.isdir(self.config.paths.repo_dir):
              %cd {self.config.paths.repo_dir}
              current_branch = subprocess.getoutput('git rev-parse --short HEAD' ).decode('utf-8').strip()
              if current_branch != self.config.commit_hash and os.path.isdir(f'{self.config.paths.repo_dir}/.git'):
                !git fetch
                !git checkout -f {self.config.commit_hash}
                self.log(f'Updated repo to tag {self.config.commit_hash}')
              else:
                self.log('Repo already up to date')
            else:
                !git clone {self.config.git_repo} {self.config.paths.repo_dir}
                %cd {self.config.paths.repo_dir}
                !git checkout -f {self.config.commit_hash}
                self.log(f'Cloned repo to tag {self.config.commit_hash}')
                !mkdir -p {self.config.paths.lib_dir}
                !wget https://github.com/camenduru/gperftools/releases/download/v1.0/libtcmalloc_minimal.so.4 -O {self.config.paths.lib_dir}/libtcmalloc_minimal.so.4

        def setup(self):
            if self.config.use_google_drive:
                from google.colab import drive
                drive.mount(self.config.paths.mount_dir)
            self.clone_repo_if_needed()
            FooocusColab.Downloader.download()


    @classmethod
    def start_ui(cls):
        setup = cls.Setup()
        setup.setup()
        %env LD_PRELOAD={self.config.paths.lib_dir}/libtcmalloc_minimal.so.4
        %env SKIP_DOWNLOAD_MODELS=true
        !python launch.py --share


FooocusColab.start_ui()
```
